```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA

inputDir = "data/raw_data/experiments/"
data = pd.read_csv(inputDir + "MFI_table_Fig7.csv")

np.random.seed(128)

data
```

```python
antigens = [
 'PDL1',
 'Ly6G',
 'CD40',
 'IL4R',
 'CD69',
 'LD',
 'CD14']

scaled = StandardScaler().fit_transform(data[antigens])

pca = PCA(n_components=3)
X_pca = pca.fit_transform(scaled) 

data["PC1"] = X_pca[:,0]
data["PC2"] = X_pca[:,1]
data["PC3"] = X_pca[:,2]

data[antigens] = MinMaxScaler().fit_transform(scaled)
data.to_csv("data/processed/MFI_scaled_pca.csv")
```
